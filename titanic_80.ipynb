```
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
```

```
datas = pd.concat([train_df, test_df])
datas = datas.drop(["PassengerId", "Survived"], axis = 1)
datas
```

```
# Step1. 補缺失值
s = datas.isna().sum()
# 篩選 series[[True....]]
s[s > 0]
```

```
most = datas["Embarked"].value_counts().idxmax()
datas["Embarked"] = datas["Embarked"].fillna(most)
```

```
# 補數字類型
med = datas.median().drop(["Pclass"])
datas = datas.fillna(med)
datas.isna().sum()
```

```
#step2. One-Hot(類別做這個)
#沒有大小關係的必做 one-Hot
#雖然Pclass屬於類型，而在補值的時候也是填上最多的，Pclass是有大小關係的

# 先準備一下我要保留的list
def name(n):
    n = n.split(",")[-1].split(".")[0]
    return n.strip()
v = datas["Name"].apply(name).value_counts()
white = v.index[:4]
```

```
def name(n): 
    n = n.split(",")[-1].split(".")[0]
    n = n.strip()
    if n in white:
        return n
    else:
        return None
datas["Name"] = datas["Name"].apply(name)
datas
```

```
v = datas["Ticket"].value_counts()
def ticket(t):
  global v
  return v[t]
datas["Ticket"] = datas["Ticket"].apply(ticket)
```
