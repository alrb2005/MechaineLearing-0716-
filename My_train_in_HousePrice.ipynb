```
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")

datas = pd.concat([train_df, test_df])
datas = datas.drop(["Id","SalePrice"], axis=1)
datas
```

```
datas = pd.get_dummies (datas)
```

```
#補缺失
s = datas.isna().sum()
s[s>0]
```

```
med = datas.median()
datas = datas.fillna(med)
datas.isna().sum()
```

```
x_train = datas.iloc[:len(train_df)]
x_predict = datas.iloc[len(train_df):]
y_train = train_df["SalePrice"]
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
reg = RandomForestRegressor(random_state=0)
params = {
    "n_estimators":range(20, 100),
    "max_depth":range(5, 10)
}
search = GridSearchCV(reg, params, cv=10, n_jobs=-1, scoring="r2")
search.fit(x_train, y_train)
print(search.best_params_)
print(search.best_score_)
```

```
reg = RandomForestRegressor(n_estimators=78, max_depth=9)
reg.fit(x_train, y_train)
pre = reg.predict(x_predict)
df = pd.DataFrame({
    "Id":test_df["Id"],
    "SalePrice":pre
})
df.to_csv("SalePrice2.csv", encoding="utf-8", index=False)
df
```
